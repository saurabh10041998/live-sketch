```python
import cv2
input = cv2.imread('./images/input.jpg')
B,G,R = cv2.split(input)
cv2.imshow("Red", R)
cv2.imshow("Blue", B)
cv2.imshow("Green", G)
cv2.waitKey(0)
merged = cv2.merge([B,G,R])
cv2.imshow("merged", merged)
cv2.waitKey(0)
cv2.imshow("blue tinted", cv2.merge([B+100, G, R]))
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import numpy as np

zeros = np.zeros(input.shape[:2], dtype = 'uint8')

cv2.imshow('red tinted', cv2.merge([zeros, zeros, R]))
cv2.imshow('green tinted', cv2.merge([zeros, G, zeros]))
cv2.imshow('blue tinted', cv2.merge([B, zeros, zeros]))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
from matplotlib import pyplot as plt
image = cv2.imread('./images/input.jpg')
histogram = cv2.calcHist([image], [0], None, [256], [0, 256])
plt.hist(image.ravel(), 256, [0,256])
plt.show()

color = ('b','g','r')

for i,col in enumerate(color):
    histogram2 = cv2.calcHist([image], [i], None, [256], [0,256])
    plt.plot(histogram2, color = col)
    plt.xlim([0,256])
```

```python
import numpy as np
import cv2
image = np.zeros((512, 512, 3), np.uint8)


#line 
cv2.line(image, (0,0), (511,511), (255,127,0), 5)
cv2.imshow('line', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#rectangle
image = np.zeros((512,512,3), np.uint8)
cv2.rectangle(image, (100,100), (300,250), (127,50,127), -1)
cv2.imshow('rectangle', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#circle
image = np.zeros((512, 512, 3), np.uint8)
cv2.circle(image, (350,350), 100, (15,75,50), -1)
cv2.imshow('gola', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#polygons
image = np.zeros((512, 512, 3), np.uint8)
pts  = np.array([[10,50], [400,50], [90,200], [50,100]], np.int32)
pts = pts.reshape((-1, 1, 2))
cv2.polylines(image, [pts], True, (0, 0, 255), 3)
cv2.imshow('polygon', image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#putting some text 
image = np.zeros((512, 512,3), np.uint8)

cv2.putText(image, 'Hello World',(75,290), cv2.FONT_HERSHEY_COMPLEX, 2, (100,170,0), 3)

cv2.imshow('text on image', image)

cv2.waitKey()

cv2.destroyAllWindows()
```
